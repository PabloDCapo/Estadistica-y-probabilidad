## Ejercicio 1


Un fabricante sabe que en promedio el 20% de las piezas electrónicas que vende requerirán reparaciones durante el primer año. En un día, vende 6 de dichas piezas.
1.   Determine la probabilidad de que durante un año no tenga reclamos de dichas piezas. Cual es la probabilidad de que reciba reclamos de todas las piezas?
2.   ¿Cuál es la probabilidad de que al menos 2 piezas tengan que ser reparadas en el primer año?
3.   Calcule la media y varianza. Interprete los resultados.

## Resolución ejercicio 1

###**Razonamiento**
La distribución binomial se utiliza cuando hay exactamente dos resultados mutuamente excluyentes de una prueba (en este caso, una pieza requiere reparación o no). Estos resultados se etiquetan como "éxito" y "fracaso".
Los parámetros de la distribución binomial son n y p donde n es el número total de pruebas (en este caso, el número de piezas vendidas) y p es la probabilidad de éxito en una prueba dada (en este caso, la probabilidad de que una pieza requiera reparación).

###Operaciones, Resultados y Gráfica

```python
# Importación de Librerías y declaración de Datos y Variables

import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np

# Parámetros de la distribución binomial
n = 6     # número de ensayos
p = 0.2   # probabilidad de éxito

# a. Probabilidad de que no haya reclamos de las piezas en un año
prob_no_reclamos = binom.pmf(0, n, p)

# b. Probabilidad de que todas las piezas requieran reparación
prob_todas_reparacion = binom.pmf(n, n, p)

# c. Probabilidad de al menos 2 piezas requieran reparación
prob_almenos_2 = 1 - binom.cdf(1, n, p)

# d. Media y varianza de la distribución binomial
media = n * p
varianza = n * p * (1 - p)

# Mostramos los resultados obtenidos.
print(f"Probabilidad de no tener reclamos en un año: Num: {prob_no_reclamos:.4f}, Porc: {prob_no_reclamos:.1%}")
print(f"Probabilidad de reclamos en todas las piezas: {prob_todas_reparacion:.4f}")
print(f"Probabilidad que al menos 2 piezas requieran reparación: Num: {prob_almenos_2}, Porc: {prob_almenos_2:.1%}")
print(f"La Media es: Num: {media}, {media:.4} la Varianza: Num: {varianza}, {varianza:.4}\n")

# Gráfica
# Valores posibles de X (número de piezas que necesitan reparación)
X = np.arange(0, n+1)

# Probabilidades asociadas a cada valor de X
P_X = binom.pmf(X, n, p)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.bar(X, P_X, color='lightblue', edgecolor='blue', alpha=0.7)
plt.xlabel('Número de piezas que necesitan reparación')
plt.ylabel('Probabilidad')
plt.title('Distribución binomial de piezas que necesitan reparación', y=1.05)
plt.xticks(X)
plt.grid(True)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('grey')
plt.gca().spines['left'].set_color('grey')

# Etiquetas en las barras
for i in range(len(X)):
    plt.text(X[i], P_X[i]+0.02, f'P({X[i]}) = {P_X[i]:.3f}', ha='center', va='bottom')

# Linea vertical para la media
plt.axvline(media, color='red', linestyle='--', label=f'Media = {media:.1f}')
plt.legend()
plt.show()
```

### **Conclusiones**
Aquí están los resultados de los cálculos:

1. La probabilidad de que no haya reclamos en un año es de aproximadamente 0.262 o 26.2%.
La probabilidad de que se reciban reclamos de todas las piezas es extremadamente baja, aproximadamente 0.0001 o 0.01%.

2. La probabilidad de que al menos 2 piezas necesiten reparación en el primer año es de aproximadamente 0.345 o 34.5%.

3. La media es 1.2000000000000002 y la varianza es 0.9600000000000002.

*Interpretación:* La media de 1.2 significa que, en promedio, se espera que 1.2 piezas necesiten reparación en el primer año.  En el gráfico está indicada con una línea punteada color rojo.
La varianza de 0.96 indica la variabilidad de los resultados alrededor de la media. En este caso, muestra que hay una variabilidad considerable en el número de piezas que podrían necesitar reparación.  Este número representa que la probabilidad puede variar casi en una pieza.


## Ejercicio 2

De 20 empleados de una empresa, 5 son técnicos. Se selecciona al azar 3 empleados para dar charlas de asesoría.
1. ¿cuál es la probabilidad de 2 empleados sean técnicos y uno no lo sea?  








## Resolución ejercicio 2(a)


### **Razonamiento**
Para resolver este ejercicio necesitamos calcular el número de formas en que podemos seleccionar 2 técnicos
de los 5 disponibles y 1 no técnico de los 15 restantes.
Luego, dividimos eso por el número total de formas en que podemos seleccionar 3 empleados de los 20.

La fórmula para calcular las combinaciones es
C(n, k) = n! / [k!(n-k)!]

Donde n es el número total de opciones y k es el número de opciones que estamos seleccionando.


### Operaciones y Resultados

```python
# Importación de Librerías y declaración de Datos y Variables
import math

n_empleados_total = 20
n_empleados_tecnicos = 5
n_empleados_no_tecnicos = n_empleados_total - n_empleados_tecnicos
n_empleados_selec = 3
n_empleados_tecnicos_selec = 2
n_empleados_no_tecnicos_selec = n_empleados_selec - n_empleados_tecnicos_selec

# Cálculo de la combinación Técnicos y No Técnicos
combinacion_empleados_tecnicos = math.comb(n_empleados_tecnicos, n_empleados_tecnicos_selec)
combinacion_empleados_no_tecnicos = math.comb(n_empleados_no_tecnicos, n_empleados_no_tecnicos_selec)

# Cálculo de la combinacion total
combinacion_total = math.comb(n_empleados_total, n_empleados_selec)

# Cálculo de la probabilidad
probabilidad = (combinacion_empleados_tecnicos * combinacion_empleados_no_tecnicos) / combinacion_total

print(f"Probabilidad de seleccionar dos técnicos y un no técnico es: Num:{probabilidad}, Porc:{probabilidad:.3%}")
```

### **Conclusiones**
La probabilidad de que 2 empleados seleccionados sean técnicos y uno no lo sea es 0.13157894736842105.
Esto significa que hay aproximadamente un 13.16% de probabilidad de que, al seleccionar al azar 3 empleados para dar charlas de asesoría, 2 sean técnicos y uno no lo sea.

###**Razonamiento**
La distribución hipergeométrica se utiliza para una distribución discreta, relacionada con un muestreo aleatorio y sin reemplazo.
Cada selección es clasificada como un éxito (en este caso, seleccionar a un técnico) o un fracaso (en este caso, seleccionar a un no técnico).

---


Los parámetros de la distribución hipergeométrica son:
N (es el tamaño de la población --> en este caso, el número total de empleados),
n (número de observaciones --> en este caso, el número de empleados seleccionados para dar charlas de asesoría) y
K (número total de éxitos en la población --> en este caso, el número total de técnicos).

###Operaciones y Resultados

```python
# Importación de Librerías y declaración de Datos y Variables
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
import numpy as np

N = 20 # Tamaño de la población
n = 3  # Nro de observaciones
K = 5  # Nro de exitos en la población

# Variable hipergeométrica
rv = hypergeom(N, n, K)

# Probabilidad
prob_2_tec_1_notec = rv.pmf(2)

mensaje = (
    f"Probabilidad de que 2 empleados seleccionados sean técnicos y uno no lo sea: {prob_2_tec_1_notec}, \n"
    f"o lo que es lo mismo: {prob_2_tec_1_notec:.3%}"
)
print(mensaje)
```

## Ejercicio 3

Una máquina de bebidas se regula para que sirva 200 ml por vaso. Si la cantidad de bebida se distribuye normalmente con desvió estándar de 15 ml.
1. ¿Cuál es la probabilidad de que una bebida contenga más de 224 ml?
2. Determine el porcentaje de bebidas que contendrán entre 191 ml y 209 ml.
3. ¿Cuál es la probabilidad de que una bebida contenga menos de 190 ml?
4. En todos los casos anteriores grafiqué la probabilidad como el área debajo de la curva.

##Resolución ejercicio 3

###**Razonamiento**
Este es un problema de distribución normal. La distribución normal es una de las distribuciones más comunes en estadística. Se caracteriza por dos parámetros:
la media (μ) y la desviación estándar (σ).
En este caso, la media es 200 ml y la desviación estándar es 15 ml.
1. La probabilidad de que una bebida contenga más de 224 ml es P(X > 224).
2. El porcentaje de bebidas que contendrán entre 191 ml y 209 ml es P(191 < X < 209).
3. La probabilidad de que una bebida contenga menos de 190 ml es P(X < 190).
Para calcular estas probabilidades, necesitamos convertir las cantidades de bebida a valores z utilizando la fórmula z = (X - μ) / σ. Luego, podemos usar la tabla de la distribución normal estándar para encontrar las probabilidades correspondientes.

###Operaciones, Resultados y Gráfica

```python
# Importación de Librerías y declaración de Datos y Variables
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

mu = 200
sigma = 15

# Variable Random
rv = norm(loc=mu, scale=sigma)

# a. Calculo probabilidad - >224ml
prob_mas_224 = 1 - rv.cdf(224)

# b. Calculo el porcentaje de bebidas entre 191 y 209 ml
percet_191_209 = rv.cdf(209) - rv.cdf(191)

# c. Calculo probabilidad - <190ml
prob_menos_190 = rv.cdf(190)

# d. Imagen
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)
plt.plot(x, rv.pdf(x), color='blue', label='Distribución Normal')
plt.fill_between(x, rv.pdf(x), where=(x > 224), alpha=0.5, label='P(X>224)', color='green')
plt.fill_between(x, rv.pdf(x), where=((x > 191) & (x < 209)), alpha=0.5, label='P(191<X<209)', color='orange')
plt.fill_between(x, rv.pdf(x), where=(x < 190), alpha=0.5, label='P(X<190)', color='red')

plt.axvline(224, color='green', linestyle='--', label='X=224')
plt.axvline(191, color='orange', linestyle='--', label='X=191')
plt.axvline(209, color='orange', linestyle='--')

##plt.xlim(150, 240)  # Ajusta el rango del eje x

print(f"1.  Probabilidad de bebidas con más de 224ml: Num:{prob_mas_224}, Porc:{prob_mas_224:.2%}")
print(f"2.  Probabilidad de bebidas con entre 191ml y 209ml: Num:{percet_191_209}, Porc:{percet_191_209:.2%}")
print(f"3.  Probabilidad de bebidas con menos de 190ml: Num:{prob_menos_190}, Porc:{prob_menos_190:.2%}\n")
print('4.  Gráfico Distribución Normal')

plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
plt.title('Distribución Normal Cantidad de Bebidas')
plt.xlabel('Cantidad Bebida (ml)')
plt.ylabel('Densidad Probabilidad')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('grey')
plt.gca().spines['left'].set_color('grey')
plt.show()
```

### **Conclusiones**
1. La probabilidad de que una bebida contenga más de 224 ml es 0.054799291699557995.
2. La probabilidad de bebidas que contendrán entre 191 ml y 209 ml es 0.45149376449985285.
3. La probabilidad de que una bebida contenga menos de 190 ml es 0.2524925375469229.
4. Se observa en el gráfico de la distribución normal correspondiente al ejercicio, y siendo que el eje de las X corresponde a los ml de las bebidas, que las diferentes áreas sombreadas por debajo de la curva,  representan las posibles incidencias de obtener bebidas con los ml indicados, que calculamos y detallamos previamente.

###**Interpretacion del gráfico**
Hay aproximadamente un 5.48% de probabilidad de que una bebida contenga más de 224 ml.
Alrededor del 45.15% de las bebidas contendrán entre 191 ml y 209 ml.
Hay aproximadamente un 25.25% de probabilidad de que una bebida contenga menos de 190 ml.
